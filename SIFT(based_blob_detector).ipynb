```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv_imshow
import subprocess
subprocess.call("pip install -U opencv-python".split())
```

```python
img = cv.imread('gdrive/MyDrive/MachineVision/Anyang.jpg') 
img = cv.resize(img,(640,480))
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# SIFT 방법으로 blob 검출
sift = cv.SIFT_create()
kp = sift.detect(gray,None)

img_b = cv.drawKeypoints(gray, kp, img, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv_imshow(img_b)
```

```python
# 영상 시계방향으로 90도 회전 후
img_r = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
gray = cv.cvtColor(img_r, cv.COLOR_BGR2GRAY)

# SIFT 방법으로 blob 검출
img_rb = cv.drawKeypoints(gray, kp, img_r, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv_imshow(img_rb)
```

```python
# 영상 2배로 키운 후
img_2 = cv.resize(img, None, fx=2, fy=2, interpolation = cv.INTER_CUBIC)
gray = cv.cvtColor(img_2, cv.COLOR_BGR2GRAY)

# SIFT 방법으로 blob 검출
img_2b = cv.drawKeypoints(gray, kp, img_2, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv_imshow(img_2b)
```
